## 1.1. Aggregate these logins counts based on 15-minute time intervals

### Resample the data into 15 minute bins

```python
# First index on the timestamp column
df.set_index('login_time', inplace=True)

# Add a numeric column (you need something for resampling to aggregate)
df['count'] = 1

# Resample in 15minute/H-hour/D-day/W-week and M-mont time intervals
df_15T = df.resample('15T', how='sum')
df_H = df.resample('H', how='sum')
df_D = df.resample('D', how='sum')
df_D = df_D[1:-1] # In order to get better look, we drop the first and last days in case of missings
df_W = df.resample('W', how='sum')
df_W = df_W[1:-1] # In order to get better look, we drop the first and last weeks in case of missing days
df_M = df.resample('M', how='sum')
df_M = df_M[:3] # In order to get better look, we drop the last month because of missing days

df_15T.head()
```

# Part 2 â Experiment and metrics design
The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active during the day. On weekends, there is reasonable activity in both cities. However, a toll bridge, with a two-way toll, between the two cities causes driver partners to tend to be exclusive to each city. The Ultimate managers of city operations for the two cities have proposed an experiment to encourage driver partners to be available in both cities, by reimbursing all toll costs.

   **Q:** What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?
   
   **A:** For a single KPI, a measure of the ratio of activity of a driver in the other city. For example, assume driver A from Gotham drove 100 times. Only 5 of them were in Metropolis. Driver A's ratio is then 5%. Currently, drivers in both cities have high ratios towards their primary city. Ultimate would like to move these ratios towards 50% so that the drivers are closer to evenly distributed: driving in both cities. This metric would show the experiment working or not working reasonably well.

   **Q:** Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:

       * How you will implement the experiment
       * What statistical test(s) you will conduct to verify the significance of the observation
       * How you would interpret the results and provide recommendations to the city operations team along with any caveats.
       
   **A:** Answer: We can assume data is always being collected. Thus we can measure our KPI before our experiment actually begins. It is important to know these numbers before the experiment starts, since that is what we will be comparing to. In some sense, this can be thought of as an A/B test, so we need to know the data very well before the change. Further, we also need to make sure we don't bias or pollute the data by announcing the experiment or change too prematurely. We don't want upcoming news to change behavior making it difficult for us to determine the results of the experiment.

Then we can announce the change and begin the toll reimbursement program. We would then start monitoring the new data. We would have to run the experiment for at least some amount of time. This could be a week or a month. It depends entirely on the scale of the data involved. We would use a statistical power test to estimate the data required to be able to discern any statistical significance. Then we could run a two-tailed z-test comparing sample proportions (since that is what the ratio is). This would give us a z-score and associated p-value. A low p-value (say at Î±=0.01) would potentially indicate a significant change in our KPI (good or bad). We could further calculate the confidence intervals for this to be able to strongly claim a statistical change in the KPI, if it exists.

Assuming our KPI does change, and in a positive direction, then we still have to be mindful of the results. We could see positive change, where more drivers are now in the city other than their primary. However, we have to estimate the amount of revenue that exact change has brought and compare it to the cost of the program (remember Ultimate is paying the tolls!). A positive KPI change coupled with negative revenue is perhaps not the ideal. We also mentioned that while the program can result in a statistical change in the KPI, it could lead to worse results. So the analysts need to be aware of this fact as well. Further, this ratio metric is time agnostic, ignoring weekdays, weekends, day or night. A further iteration of this analysis would also include this (by perhaps binning in time) to see if perhaps there are certain times when the program is being used more than others.

#### Checking percentage of outliers

```python
# Computing percentages of outliers with a "for loop"
data=df3.copy()
for feature in data.select_dtypes(include=[np.float64, np.int64]).columns:
    step=1.5*((np.percentile(data[feature], q=75))-(np.percentile(data[feature], q=25)))
    lower_bound=(np.percentile(data[feature], q=25))-step
    upper_bound=(np.percentile(data[feature], q=75))+step
    a = (len(list(data.loc[data[feature]<lower_bound].index))+
            (len(list(data.loc[data[feature]>upper_bound].index))))/df3.shape[0]
    print('{}:  {}'.format(feature,a))
```

#### Comment on "outliers checking":
* No need to touch Features which have almost "0" percentage of outliers
* "avg_dist" has %9 and avg_surge has %16 outlier rates. But dropping all outlier rows of these features may effect the prediction negatively. Instead of deleting all rows between lower and upper bound, we can select the values which are above a specific value and then fill them with max values of their own features.
* We will try the detect a limit which values become very rare after that. 
* As we see above outlier graphs, rare limits for features look like 
    - "55" for "avg_dist"

## Handling Skewness

### Handling Skewness - Checking Distributions
* Step-1: We will check the distibution plots
* Step-2: We will decide the features which need to be unskeewed and methods for handling unskewness
* Step-3: We will transform the dtected features to unskew teir distributions.

#### Checking DistributionsÂ¶

```python
# create the figure instance and axes
fig1, axes = plt.subplots(1,6, figsize=[16,2])

# plot distributions of all transformed features with "for loop" 
for i,clm in enumerate(df3.select_dtypes(include=[np.float64, np.int64]).columns):
    plt.subplot(1,6,i+1)
    sns.distplot(df3[clm])
    plt.ylabel(clm)

plt.tight_layout() 
plt.show()
```

#### Comments on distiribution checking:
* All features need to be unskewed
* All features have positive values but some features have many zeros.
* We will use two different methods for transformation:
    - We will apply "square root" transformation on features with many "zeros" since "0" values will return "inf" values with "log transformation"
    - We will apply "log transformation" on other features.

#### Applying Transformation

```python
df4=df3.copy()
for clm in df4.select_dtypes(include=[np.float64, np.int64]).columns:
    if df4[clm].min()==0.0:
        df4[clm] = df4[clm].apply(lambda x: x**(1/2))
    else:
        df4[clm] = np.log(df4[clm])
df4.describe()
```

#### Comparing Original and Unskewed versions

```python
# create the figure instance and axes
fig2, axes = plt.subplots(1,6, figsize=[16,2])
# plot distributions of all transformed features with "for loop" 
for i,clm in enumerate(df3.select_dtypes(include=[np.float64, np.int64]).columns):
    plt.subplot(1,6,i+1)
    sns.distplot(df3[clm])
    plt.ylabel(clm)
    plt.title('Original')

plt.tight_layout() 
plt.show()

# create the figure instance and axes
fig3, axes = plt.subplots(1,6, figsize=[16,2])
# plot distributions of all transformed features with "for loop" 
for i,clm in enumerate(df4.select_dtypes(include=[np.float64, np.int64]).columns):
    plt.subplot(1,6,i+1)
    sns.distplot(df4[clm])
    plt.ylabel(clm)
    plt.title('Unskewed')

plt.tight_layout()
plt.show()
```

#### Comments on unskewing:
* Transformation worked well on "av_dist" feature
* And trimed the high peaks of the other features but since they have very unbalanced distirbution, getting an ideal normal distribution is very difficult

# Modeling

## Preprocessing

### Creating Target Variable
* We need to create a target variable, since we are asked the predict a binary classification about whether a driver will be active at the end of sixt month.

```python
active_dict = {x:False for x in range(0, len(df4))}

#Loop over rows
for i in range(len(df4)):
    
    #Set the start date of this row
    start = df4['signup_date'][i]
    
    #Set the end date for this row: 6 months later
    end = start + pd.Timedelta('180D')
    
    #Set the time window, a month before end date
    previous = end - pd.Timedelta('30D')
    
    #Was the last trip within the last month?
    if previous <= df4['last_trip_date'][i] <= end:
        active_dict[i] = True

#What is the percentage of active users?
num_active = sum(active_dict.values())
print('There are {} ({}) active users.'.format(num_active, num_active/len(df4)))   

#Add active user info to the dataframe
df4['is_active'] = pd.DataFrame(active_dict.items(), columns=['ind', 'active'])['active']
```

### Dropping Date Variables

```python
df4.drop(['last_trip_date', 'signup_date'], axis=1, inplace=True)
df4.head()
```

## Training

```python
X = df4.drop(['is_active'], axis=1)
y = df4['is_active']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, make_scorer
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier

#Likewise, we will setup a stratified kfold strategy for our cross-validated scoring that we will do.
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

#We will also use the ROC Area Under the Curve metric to compare our models. 
score_func = make_scorer(roc_auc_score, greater_is_better=True)

#Setup the RF object
rf = RandomForestClassifier(criterion='gini', class_weight='balanced', n_jobs=-1)

#Hyperparameter grid
param_grid = {'n_estimators' : [40, 50, 60], 'min_samples_split' : [2, 3, 4], 
              'max_depth' : [4, 7, 10]}

#Setup the Gridsearch
rf_cv = GridSearchCV(rf, param_grid, cv=cv, scoring=score_func, verbose=1)

#Fit the data
rf_cv.fit(X_train, y_train)

#Print out the best model
print('Best RF Params: {}'.format(rf_cv.best_params_))
print('Best RF Score : %f' % rf_cv.best_score_)
```

#### With the best parameters, we can re-fit another RF object with those values.

```python
rf_best = RandomForestClassifier(criterion='gini', class_weight='balanced', n_jobs=-1, 
                                 n_estimators=rf_cv.best_params_['n_estimators'],
                                 min_samples_split=rf_cv.best_params_['min_samples_split'], 
                                 max_depth=rf_cv.best_params_['max_depth'])

rf_best.fit(X_train, y_train)
```

#### Now, we can find the score in the test dataset.

```python
#Print the score/metric for the input model which must be fit
def get_scores(model, name='unk', score=roc_auc_score):
    
    #Make prediction for the model
    pred = model.predict(X_test)
    pred_prob = model.predict_proba(X_test)[:,1]

    #Calculate the score on the entire test set
    auc = score(y_test, pred_prob)
    
    #Calculate the score on CV test set
    cv_scores = cross_val_score(model, X_test, y_test, cv=cv, scoring=score_func)
    mean_cv = cv_scores.mean()
    std_cv = cv_scores.std()
    
    #Save to dataframe to be used for printing only
    model_score = pd.DataFrame(columns=['Model', 'Test score', 'CV score'])
    model_score.loc[0] = [name, auc, mean_cv]
    model_score.set_index('Model', inplace=True)
    print(model_score)
```

```python
get_scores(rf_best, 'RF')
```
